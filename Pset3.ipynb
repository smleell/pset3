```python
# Problem 1

fhand = open('/blue/bsc4452/share/Class_Files/data/CO-OPS_8727520_wl.csv')
next(fhand) # Skip first line

# Set the values
max_level = None 
max_date = None
max_time = None
for line in fhand:
    lis = line.split(',') # Convert each line to list
    date = lis[0]
    time = lis[1]
    preliminary = lis[3]
    level = float(preliminary.strip('"')) # Convert string to float
    if max_level is None or level > max_level:
            max_level = level
            max_date = date
            max_time = time
print(f'The highest water level is {max_level} ft on {max_date} at {max_time}.')
```
